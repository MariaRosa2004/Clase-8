```python
print('CENTRO DE MONITOREO PARA BROTES DE ENFERMEDADES')
class CentroMonitoreo:
    def __init__(self):
        self.__personas=[]
    def VerPersonas(self):
        return self.__personas
    def CantidadPersonas(self):
        return len(self.__personas)
    def BuscarPersonas(self,NombrePersona):
        if NombrePersona in self.__personas:
            return True
        else:
            return False
    def guardarPersona(self,nombre):
        self.__personas.append(nombre)
        return nombre
    def __str__(self):
        return ','.join(self.__personas)

class personas:
    def __init__(self):
        self.__nombre=''
        self.__sintomas=''
        self.__ubicacion=''
        self.__fecha=''
    #Nombre
    def VerNombre(self):
        print (self.__nombre)
    def AsignarNombre(self):
        var=input('Ingresa nombre')
        self.__ubicacion=var
        return self.__ubicacion
    #Fecha
    def VerFecha(self):
        print (self.__fecha)
    def AsignarFecha(self):
        var=input('Ingresa fecha')
        self.__fecha=var
        return self.__fecha
    #Ubicación
    def VerUbicacion(self):
        print (self.__ubicacion)
    def AsignarUbicacion(self):
        var=input('Ingresa ubicacion')
        self.__ubicacion=var
        return self.__ubicacion
    #Sintomas
    def AsignarSintomas(self):
        while True:
          menu2=input('¿Desea agregar sintomas? 1.Si 2.No')
          if menu2=='1':
            varsin = input('Ingrese sintoma')
            self.__sintomas=varsin
            return varsin
          elif menu2=='2':
            break

    def VerSintoma(self):
        return self.__sintomas



    def __str__(self):
        return f'''Nombre: {self.__nombre}
        Ubicacion: {self.__ubicacion}
        Fecha: {self.__fecha}
        Sintomas: {self.__sintomas}'''


monitoreo=CentroMonitoreo()
persona=personas()
sintomas=[]
Centro=[]
def main():
  while True:
      menu=input('''Ingresa opcion del menu:
              1. Ver cantidad de personas
              2. Almacenar Persona
              3. Mostrar
              4. Salir
              >''')
      if menu=='1':
          print(f'La cantidad de personas ingresadas al sistema son:{monitoreo.CantidadPersonas()}')
      elif menu=='2':
          persona1=input('Ingrese persona')
          monitoreo.BuscarPersonas(persona1)
          if monitoreo.VerPersonas() is True:
            for i in monitoreo.VerPersonas():
              if i==persona1:
                print(i)
          else:
              persona.AsignarNombre()
              persona.AsignarFecha()
              persona.AsignarSintomas()
              persona.AsignarUbicacion()
              Centro.append(persona)
              print('La persona se ha almacenado exitosamente')
      elif menu=='3':
        if Centro==[]:
         print('No hay pacientes registrados aún')
        else:
          for paciente in Centro:
            print(paciente)
      elif menu=='4':
          break
      else:
          print('El numero ingresado no es correcto intente nuevamente')

if __name__ == "__main__":
  main()
```
